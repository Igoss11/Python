```python
# f(x) = sin(x)^2 - cos(x)^2

# 1. Определить корни
# 2. Найти интервалы, на которых функция возрастает
# 3. Найти интервалы, на которых функция убывает
# 4. Построить график
# 5. Вычислить вершину
# 6. Определить промежутки, на котором f > 0
# 7. Определить промежутки, на котором f < 0
```

```python
from sympy import symbols, sin, cos, Symbol, solve, diff
from sympy.plotting import plot
x = Symbol('x')
```

```python
# 1. Определить корни

solve(sin(x)**2 - cos(x)**2, x)
```

```python
# 2. Найти интервалы, на которых функция возрастает

ans = diff(sin(x)**2 - cos(x)**2, x)

solve(ans>0)
```

```python
# 3. Найти интервалы, на которых функция убывает

solve(ans<0)
```

```python

# 4. Построить график
x = Symbol('x')
func = plot(sin(x)**2-cos(x)**2, show=False)
func.show()
```

```python
# 5. Вычислить вершину

solve(ans)
```

```python
# 6. Определить промежутки, на котором f > 0

solve(sin(x)**2 - cos(x)**2 > 0, x)
```

```python
# 7. Определить промежутки, на котором f < 0

solve(sin(x)**2 - cos(x)**2 < 0, x)
```
